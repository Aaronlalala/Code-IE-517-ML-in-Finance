## Yield Curve

```python
def yield_curve_snapshot(date):
    plt.plot(df.loc[df["Date"]==date].drop(columns=["Adj_Close"]).values[0][1:])
    plt.xlabel("Maturity Years")
    plt.ylabel("Yield")
    plt.title("Yield Curve at " + date)
    plt.show()
    
```

# Principal Component Analysis 

```python
# Since we want to see the time series data, need to sort x in time order
X_train["Date Time"] = pd.to_datetime(X_train["Date"])
X_train = X_train.sort_values(by="Date Time", ascending=True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
# Since all features are yields (in the same scale)
pca.fit(X_train.drop(columns=["Date", "Date Time"]))
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pca.components_[0])
plt.plot(pca.components_[1])
plt.plot(pca.components_[2])
plt.legend(["general interest level", "Steepness of yield curve", "Curvature of yield curve"])
plt.show()
```

Large steepness make the second pc more negative.

# Fit Predictive Models
Genearlly, we can't directly use supervised learning model for timeseries data, because it assume i.i.d. distribution of features. For the purpose of assignment, we just believe it's true.

## Linear Regression Model

```python
# since we sorted the data for visualization of PCA, we need to split again
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Adj_Close"]), df["Adj_Close"], random_state=42, train_size=0.85)
```

### Fit Untransformed Data

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
lm = LinearRegression().fit(X_train.drop(columns=["Date"]), y_train)

y_train_pred = lm.predict(X_train.drop(columns=["Date"]))
y_pred = lm.predict(X_test.drop(columns=["Date"]))

# R2 score
print("Training r2_score is ", r2_score(y_train_pred, y_train))
# RMSE
print("Training RMSE is", mean_squared_error(y_train_pred, y_train)**0.5)

# R2 score
print("Testing r2_score is ", r2_score(y_pred, y_test))
# RMSE
print("Testing RMSE is", mean_squared_error(y_pred, y_test)**0.5)
```

### Fit Transformed Data

```python
pca = PCA(n_components=3)
# Since all features are yields (in the same scale)
X_train_transformed = pca.fit_transform(X_train.drop(columns=["Date"]))
X_test_transformed = pca.fit_transform(X_test.drop(columns=["Date"]))

lm2 = LinearRegression().fit(X_train_transformed, y_train)
y_train_pred = lm2.predict(X_train_transformed)
y_pred = lm2.predict(X_test_transformed)

# R2 score
print("Training r2_score is ", r2_score(y_train_pred, y_train))
# RMSE
print("Training RMSE is", mean_squared_error(y_train_pred, y_train)**0.5)

# R2 score
print("Testing r2_score is ", r2_score(y_pred, y_test))
# RMSE
print("Testing RMSE is", mean_squared_error(y_pred, y_test)**0.5)
```

After applying PCA for high dimension and highly correlated features data, we could significantly increase the linear model prediction.

## SVM

### Linear Kernel

```python
# Import linear support vector regressor
from sklearn.svm import LinearSVR
# Split data again to make sure data is not manipulated by mistakes previously.
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Adj_Close", "Date"]), df["Adj_Close"], random_state=42, train_size=0.85)

pca = PCA(n_components=3)
X_train_transformed = pca.fit_transform(X_train)
X_test_transformed = pca.transform(X_test)
```

```python
# Fit LinearSVR to the original data and transformed data
svm_original = LinearSVR( max_iter=3000)
svm_original.fit(X_train, y_train)


# Fit LinearSVR to the original data and transformed data
svm_transformed = LinearSVR( max_iter=3000)
svm_transformed.fit(X_train_transformed, y_train)
```

```python
y_train_pred_original = svm_original.predict(X_train)
y_train_pred_transformed = svm_transformed.predict(X_train_transformed)

y_test_pred_original = svm_original.predict(X_test)
y_test_pred_transformed = svm_transformed.predict(X_test_transformed)
```

```python
# Evaluate the performance using R2 score and RMSE for both original and transformed data
r2_train_original = r2_score(y_train, y_train_pred_original)
r2_test_original = r2_score(y_test, y_test_pred_original)

r2_train_transformed = r2_score(y_train, y_train_pred_transformed)
r2_test_transformed = r2_score(y_test, y_test_pred_transformed)

rmse_train_original = np.sqrt(mean_squared_error(y_train, y_train_pred_original))
rmse_test_original = np.sqrt(mean_squared_error(y_test, y_test_pred_original))

rmse_train_transformed = np.sqrt(mean_squared_error(y_train, y_train_pred_transformed))
rmse_test_transformed = np.sqrt(mean_squared_error(y_test, y_test_pred_transformed))

# Print out the results
print("Results for original data:")
print("R2 score for training set:", r2_train_original)
print("R2 score for test set:", r2_test_original)
print("RMSE for training set:", rmse_train_original)
print("RMSE for test set:", rmse_test_original)

print("\nResults for transformed data:")
print("R2 score for training set:", r2_train_transformed)
print("R2 score for test set:", r2_test_transformed)
print("RMSE for training set:", rmse_train_transformed)
print("RMSE for test set:", rmse_test_transformed)
```

SVM with linear kernel will not converge, it may indicate that the relationship between the input variables and the target variable is not linear.

### RBF Kernel

```python
# Predict the target values for both the training and testing data for both original and transformed data
y_train_pred_original = svm_original.predict(X_train)
y_train_pred_transformed = svm_transformed.predict(X_train_transformed)

y_test_pred_original = svm_original.predict(X_test)
y_test_pred_transformed = svm_transformed.predict(X_test_transformed)
```

We could see that results of nonlinear kernel significantly better than linear kernel.
